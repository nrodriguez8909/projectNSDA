## Random networks

```python
# paquetes

# import the relevant libraries
import networkx as nx
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
# funciones

# tomado de https://gist.github.com/jonathondilworth/bca4cd6a52d0f6e58874

# our own fancy function for producing a random value to implement stochastic-ism
def rand(): # produces a random value between 0.0 and 1.0
	random_number = random.randint(0,10)
	f_rand = float(random_number)
	d_rand = f_rand * 0.1
	return d_rand

# the predicted number of edges: p x N x (N - 1) / 2
def predicted_number_of_edges(p, N):
	return (p * N * (N - 1)) / 2
```

```python
# simular

# ejemplo 1: acá se va a asumir todo conectado con todo, por eso p = 1

# probability p
p = 1.0

# number of nodes
n = 12

def graficar_red(p,n):
  # lets predict the number of edges
  print("Predicted number of edges: " + str(predicted_number_of_edges(p, n)))

  # our graph
  g = nx.Graph()

  # start with n isolated nodes
  for x in range (0, n):
  	g.add_node(x)

  # consider every possible link between each node
  for x in range (0, n):
  	# we disregard the last node, because this link has already been considered (#1)
  	for y in range (x, n):
  		# add it with probability p, unless p=1, in which case add it anyway
  		# ensuring the edge is not being added to itself
  		if (rand() < p or p == 1.0) & (x != y):
  			# we do not need to check to see if the edge already exists, because of #1
  			g.add_edge(x, y)

  print("Actual number of edges: " + str(g.number_of_edges()))

  # lets draw this lovely thing!
  nx.draw(g, pos=nx.circular_layout(g))
  plt.show()
  return(g)

graficar_red(p,n)
```

## Combinar los 3 modelos con nuestra red

```python
#aplicando a nuestro caso
n = 3783
p = 2 * 24186 / (n * (n - 1) )
k = 4 # este valor no se cual poner
m = 2 # este valor no se cual poner

n = 3783
p_sw = 2 * (24186*100  ) / (n * (n - 1) )
k_sw = 4 

# modelos
G_random = nx.erdos_renyi_graph( n = n , p = p  )
G_smallWord = nx.watts_strogatz_graph(n, k_sw, p_sw)
G_preferential = nx.barabasi_albert_graph(n, m)

# Calcula el grado de centralidad de los nodos

centrality_random = nx.degree_centrality(G_random)
centrality_smallWord = nx.degree_centrality(G_smallWord)
centrality_preferential = nx.degree_centrality(G_preferential)

#densidades
sns.distplot(list(centrality_random.values()), hist=True, kde=True, color='blue', 
             hist_kws={'edgecolor':'black'}, kde_kws={'linewidth': 2})
sns.distplot(list(centrality_smallWord.values()), hist=True, kde=True, color='black', 
             hist_kws={'edgecolor':'black'}, kde_kws={'linewidth': 2})
sns.distplot(list(centrality_preferential.values()), hist=True, kde=True, color='red', 
             hist_kws={'edgecolor':'black'}, kde_kws={'linewidth': 2})

# Agrega etiquetas y título a la gráfica
plt.xlabel('Grado de centralidad')
plt.ylabel('Densidad')
plt.title('Densidad del grado de centralidad de los nodos')

# limites de grafico
plt.xlim(0,0.01)

# Muestra la gráfica
plt.show()
```
